```
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('Pokemon.csv', index_col = 0, encoding='unicode-escape')

df.head()
```

```
sns.lmplot(x='Attack', y='Defense', data=df)
plt.show()
```

```
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg=False, hue='Stage')
plt.show()
```

```
df_copy = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
sns.boxplot(data=df_copy)
```

```
sns.violinplot(data=df_copy)
plt.show()
```

```
sns.violinplot(x='Type 1', y='Attack', data=df)
plt.show()
```

```
plt.figure(figsize=(10,6))
sns.violinplot(x='Type 1', y='Attack', data=df)
plt.show()
```

```
corr = df_copy.corr()
sns.heatmap(corr)
```

```
sns.distplot(df.Attack, color='blue') #this is a distplot which is also known as a histogram
```

```
sns.countplot(x='Type 1', data=df)
plt.xticks(rotation=-45)
```

```
sns.jointplot(df.Attack, df.Defense, kind='kde', color='lightblue') #kde indicated that we want a density plot.
```
